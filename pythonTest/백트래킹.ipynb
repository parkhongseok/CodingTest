```python
def backtrack(N, M, sequence, used):
    # M개의 수를 다 고르면 결과 출력
    if len(sequence) == M:
        print(*sequence)
        return
    
    for i in range(1, N + 1):
        if used[i]:  # 이미 선택한 숫자는 skip
            continue

        # 상태 변경: 숫자 i를 수열에 추가
        sequence.append(i)
        used[i] = True

        # 재귀 호출: 선택한 숫자 i를 추가한 뒤, 다음 깊이로 이동
        backtrack(N, M, sequence, used)

        # 되돌리기: 마지막에 선택한 숫자 i를 수열에서 제거
        sequence.pop()
        used[i] = False

# 입력
N, M = map(int, input().split())
sequence = []  # 현재 수열
used = [False] * (N + 1)  # 각 숫자 사용 여부를 체크

backtrack(N, M, sequence, used)
```

```python
import sys
from io import StringIO

test1 = """3 1
4 4 2"""

test2 = """4 2
9 7 9 1"""
```

```python
sys.stdin = StringIO(test1)

import sys
from itertools import combinations

ipt = sys.stdin.readline

N, M = map(int, ipt().split())
NumberList = sorted([int(i) for i in ipt().split()])

# 가능한 인덱스 조합 모두 생성
index_combinations = combinations(range(N), M)
result_set = set()

for comb in index_combinations:
    picked = [NumberList[i] for i in comb]
    if picked == sorted(picked):  # 오름차순 조합만
        result_set.add(tuple(picked))

# 정렬해서 출력
for item in sorted(result_set):
    print(*item)
```
