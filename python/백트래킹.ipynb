자연수 N과 M이 주어졌을 때, 아래 조건을 만족하는 길이가 M인 수열을 모두 구하는 프로그램을 작성하시오.

1부터 N까지 자연수 중에서 중복 없이 M개를 고른 수열

```python
def backtrack(N, M, sequence, used):
    # M개의 수를 다 고르면 결과 출력
    if len(sequence) == M:
        print(*sequence)
        return
    
    for i in range(1, N + 1):
        if used[i]:  # 이미 선택한 숫자는 skip
            continue

        # 상태 변경: 숫자 i를 수열에 추가
        sequence.append(i)
        used[i] = True

        # 재귀 호출: 선택한 숫자 i를 추가한 뒤, 다음 깊이로 이동
        backtrack(N, M, sequence, used)

        # 되돌리기: 마지막에 선택한 숫자 i를 수열에서 제거
        sequence.pop()
        used[i] = False

# 입력
N, M = map(int, input().split())
sequence = []  # 현재 수열
used = [False] * (N + 1)  # 각 숫자 사용 여부를 체크

backtrack(N, M, sequence, used)
```
