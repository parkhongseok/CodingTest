```python
# 백트레킹 방식 사용
def backtrack(N, M, sequence, used):
    # M개의 수를 다 고르면 결과 출력
    if len(sequence) == M:
        print(*sequence)
        return
    
    for i in range(1, N + 1):
        if used[i]:  # 이미 선택한 숫자는 skip
            continue
        
        # 상태 변경: 숫자 i를 수열에 추가
        sequence.append(i)
        used[i] = True
        
        # 재귀 호출: 선택한 숫자 i를 추가한 뒤, 다음 깊이로 이동
        backtrack(N, M, sequence, used)
        
        # 되돌리기: 마지막에 선택한 숫자 i를 수열에서 제거
        sequence.pop()
        used[i] = False

# 입력
N, M = map(int, input().split())
sequence = []  # 현재 수열
used = [False] * (N + 1)  # 각 숫자 사용 여부를 체크

backtrack(N, M, sequence, used)
```

```python
# 이터레이터 툴 사용한 조합 구하기
from itertools import combinations
N, M = map(int, input().split())
for items in combinations(range(1, N+1), M):
  print(*items)
```

```python
N, M = map(int, input().split())

def dfs(start = 1, answer = []):
  if len(answer) == M:
    print(*answer)
    return
  
  for i in range(start, N+1):
    answer.append(i)
    dfs(i + 1, answer)
    answer.pop()
dfs()
```
